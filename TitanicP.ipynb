##Titanic Prediction

```python
import pandas as pd

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

from google.colab import drive
```

```python
drive.mount("/content/gdrive")
```

```python
df_gender_submission = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/gender_submission.csv')
df_train = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/train.csv')
df_test = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/test.csv')
```

```python
df_train.head()
```

#Random Forest

```python
df_gender_submission['Survived'].value_counts()
```

```python
y = df_train['Survived']
y
```

```python
features = ['Pclass','Sex','SibSp','Parch']

X = pd.get_dummies(df_train[features])
X_test = pd.get_dummies(df_test[features])
```

```python
model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X,y)
```

```python
predictions = model.predict(X_test)
```

```python
df_gender_submission_array = df_gender_submission['Survived'].to_numpy()
```

```python
accuracy = accuracy_score(df_gender_submission_array, predictions)
print("Acurácia:", round(accuracy*100,2) , "%")
```

```python
output = pd.DataFrame({'PassengerId': df_test.PassengerId, 'Survived': predictions})
output.to_csv('submission.csv', index=False)
print("Your submission was successfully saved!")
```
